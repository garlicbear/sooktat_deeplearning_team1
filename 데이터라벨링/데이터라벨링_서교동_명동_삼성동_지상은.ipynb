```python
=>중복 제거
=>리뷰 없는 데이터 삭제
=>뒤에 붙은 ‘더보기’, ‘접기’ 삭제
=>영어 리뷰 삭제

-1, 2점: 무조건 부정/ 4, 5점: 무조건 긍정
3점: (리뷰별점-평균별점)>=1이면 긍정, 아니면 부정
리뷰가 1개이면서 리뷰별점이 3점인 사람은 무조건 날리기
평균별점이 2점 미만인 사람은 무조건 날리기


1. 오늘 밤까지 크롤링 완료
-윤서 님: 성수1동, 성수2동, 여의도동,
-상은 님: 명동, 삼성동, 서교동
-소연 님: 종로 1~4가동, 서초동, 청담동
-희진 님: 논현동, 가산동, 역삼동
2. 크롤링할 때 평균 별점, 리뷰 개수도 크롤링하기. 
3. 전처리 방식
-중복 제거/ 리뷰 없는 데이터 제거/ 뒤에 ‘더보기’, ‘접기’ 삭제/ 영어 리뷰 삭제
-악성댓글러(평균평점이 2점 미만인 경우) 삭제
```

```python
import pandas as pd
import numpy as np
```

```python
sg=pd.read_csv("total_seogyodong.csv")
sg.info()
```

```python
my=pd.read_csv("total_myeongdong.csv")
my.info()
```

```python
ss=pd.read_csv("total_samseongdong.csv")
ss.info()
```

```python
total=pd.concat([sg,my,ss],ignore_index=True)
total
```

# 중복, (더보기&접기) 공백 제거

```python
total=total.drop_duplicates() #중복 제거
total=total[total['review']!="더보기"] #리뷰없는 데이터 제거
total['review']=total['review'].str.replace('더보기','') #더보기 제거
total['review']=total['review'].str.replace('접기','') #접기 제거
total=total[total['review']!=""]
total
```

# 영어 포함된 리뷰 제거

```python
total['review'] = total['review'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","",regex=True) 
total=total[total['review']!=""]
total
```

# 전처리 완료

```python
total
```

# 긍/부정 판별

```python
label=[
    (total['star']>=4),
    (total['star']<=2),
    (total['star']==3) & (total['star']-total['ave_star']>1.5),
    (total['star']==3) & (total['star']-total['ave_star']<-1.5),
    (total['star']==3) & (total['ave_star']-total['star']>=1.5),
    (total['star']==3) & (total['ave_star']-total['star']<=1.5),]

labelling=[0,1,0,1,2,2]

total['label']=np.select(label,labelling,default='Not specified')
```

```python
total=total[total['label']!="2"]
total
```
