```python
import pandas as pd
```

```python
gs = pd.read_csv("total_review_gs_raw.csv")
nh = pd.read_csv("total_review_nh_raw.csv")
ys = pd.read_csv("total_review_ys_raw.csv")
```

```python
total_review = pd.concat([gs, nh, ys], ignore_index=True)
```

```python
total_review
```

```python
total_review.to_csv("가산동_논현동_역삼동_원본데이터셋_김희진.csv",encoding="utf-8-sig" ,index=False)
```

```python
df_copy = total_review.copy()
```

```python
df_copy = df_copy.drop_duplicates(['review'], keep = 'first') #중복제거
df_copy
```

```python
df_copy['review'] = df_copy['review'].str.replace('더보기','') 
#뒤에 붙은 '더보기' 제거
```

```python
df_copy['review'] = df_copy['review'].str.replace('접기','')
#뒤에 붙은 '접기' 제거
```

```python
df_copy['review'] = df_copy['review'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")
#한글과 띄어쓰기 제외하고 모두 제거
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
df_copy
```

```python
import numpy as np
```

```python
df_copy['review'] = df_copy['review'].str.replace('^ +', "") # white space 데이터를 empty value로 변경
df_copy['review'].replace('', np.nan, inplace=True)
print(df_copy.isnull().sum())
```

```python
df_copy.loc[df_copy.review.isnull()][:5]
```

```python
df_copy = df_copy.dropna(how = 'any')
```

```python
df_copy
```

```python
df_copy['ave_star'] = pd.to_numeric(df_copy['ave_star'])
```

```python
#긍정부정 라벨링
```

```python
df_copy_drop = df_copy.drop(df_copy[(df_copy['star'] == 3) & (abs(df_copy['star']-df_copy['ave_star']) <= 1.5)].index)
#|(평균 평점-리뷰 평점)|≤1.5이면서 3점인 리뷰는 모두 삭제하기. 
```

```python
df_copy_drop
```

```python
label = []
for i, j in zip(df_copy_drop['star'], df_copy_drop['ave_star']):
    if i >= 4:
        label.append('0')
    elif i <= 2:
        label.append('1')
    else:
        if abs(i-j) < -1.5:
            label.append('1')
        else:
            label.append('0')
```

```python
df_copy_drop['label'] = label
```

```python
df_copy_drop
```

```python
df_copy_drop.to_csv("가산동_논현동_역삼동_최종데이터셋_김희진.csv",encoding="utf-8-sig" ,index=False)
```
