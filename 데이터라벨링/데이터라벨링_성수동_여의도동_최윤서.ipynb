## 데이터 불러오기

```python
import pandas as pd
```

```python
sung1_df=pd.read_csv("total_review_성수1동_후기개수버전.csv")
```

```python
sung1_df.info()
```

```python
sung2_df=pd.read_csv("total_review_성수2동_후기개수버전.csv")
```

```python
sung2_df.info()
```

```python
yo_df=pd.read_csv("total_review_여의도.csv")
```

```python
yo_df.info()
```

## 데이터 합치기

```python
df=pd.concat([sung1_df,sung2_df,yo_df],ignore_index=True)
```

```python
#df.to_csv("성수동_여의도동_원본데이터셋_최윤서.csv",index=False, encoding='utf-8-sig') #데이터 저장
```

## 중복제거

```python
df
```

```python
df_1=df.drop_duplicates()
```

## 리뷰 없는 데이터 제거

```python
df_2=df_1[df_1["review"]!="더보기"]
```

```python
df_2
```

```python
df_3=df_2.copy()
```

## 뒤에 "더보기", "접기" 삭제

```python
df_3["review"].replace('더보기$',"",inplace=True,regex=True)
```

```python
df_3["review"].replace('접기$',"",inplace=True,regex=True)
```

```python
df_3
```

```python
df_4=df_3.copy()
```

## 영어 리뷰 삭제

```python
df_4['review'] = df_4['review'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","") #한글만 남기기.
```

```python
df_5=df_4.copy() #df_4: 영어 리뷰 삭제
```

## 공백으로만 이루어진 리뷰 제거

```python
df_5["review"].replace('\s*$',"",inplace=True,regex=True) #공백으로만 이루어진 거 제거
```

```python
df_5=df_5[~(df_5["review"]=="")] #리뷰 길이가 0인 거 제거.
```

```python
df_5=df_5.drop_duplicates() #중복 제거
```

```python
df_5["positive?"]=""
df_5.loc[df_5.star>=4,"positive?"]=0 #4, 5점은 긍정으로
df_5.loc[df_5.star<=2,"positive?"]=1 #1, 2점은 부정으로
df_5=df_5.loc[~((df_5.star==3) & (abs(df_5.star-df_5.ave_star)<=1.5)),:] #3점 중 리뷰 평점이랑 평균 평점의 차이가 1.5 이하인 건 삭제
df_5
```

```python
df_5.loc[(df_5.star-df_5.ave_star)>1.5,"positive?"]=0
df_5.loc[(df_5.star-df_5.ave_star)<-1.5,"positive?"]=1
df_5
```

```python
df_5=df_5.reset_index(drop=True) #인덱스 바꿔주기
```

```python
df_5.to_csv("최종데이터셋_성수동_여의도동_최윤서.csv",encoding='utf-8-sig',index=False)
```
